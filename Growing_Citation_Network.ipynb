```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_web = pd.read_csv('webpages.csv', index_col=0)
```

```python
df_web
```

```python
x_web = np.array(df_web.index)
y_web = np.array(df_web['webpages'])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.xticks(x_web[::2])
ax.ticklabel_format(style='plain', axis="y")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.spines["right"].set_color("none") 
ax.spines["left"].set_color("none") 
ax.spines["top"].set_color("none") 
ax.set_ylabel('Total Number of Webpages',fontsize=12)
ax.set_xlim([1990,2019])
plt.bar(x_web, y_web, width=0.6 ,color="gray", edgecolor='#000000', label="Total number of webpages")
#plt.legend(loc="upper left", fontsize=12)
plt.savefig("tn_webpages.png", format = 'png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
